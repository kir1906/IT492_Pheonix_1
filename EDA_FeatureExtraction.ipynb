```python
# Categorizing status of a business into 'Open', 'Temporarily Closed', 'Permanently Closed', or 'Unknown'
def categorize_status(state):
    if pd.isna(state):
        return 'Unknown'
    elif 'open' in state.lower() or 'reopens' in state.lower():
        return 'Open'
    elif 'temporarily closed' in state.lower():
        return 'Temporarily Closed'
    elif 'permanently closed' in state.lower():
        return 'Permanently Closed'
    
business_info['status'] = business_info['state'].apply(categorize_status)
business_counts = business_info['status'].value_counts()
# Plot to show distribution of availability of businesses
plt.figure(figsize=(6, 6))
plt.pie(business_counts, labels=business_counts.index, autopct='%1.1f%%', colors=sns.color_palette('magma', len(business_counts)), textprops={'color': 'white'})
plt.title('Distribution of availability of businesses')
plt.legend(bbox_to_anchor=(1, 1), loc="upper left")  
plt.show()
```

### *Data pre-processing and feature extraction on metadata dataset containing information about businesses*

```python
# Removing non-essential columns
business_info.drop(axis=0, inplace=True, columns=['address', 'latitude', 'longitude', 'url', 'hours'])
```

```python
# Filtering out permanently closed businesses and then dropping the state of a business column
business_info = business_info[business_info.state != 'Permanently closed']
business_info.drop(axis=0, inplace=True, columns=['state'])
```

```python
# Fills missing values with an empty string
business_info['category'] = business_info['category'].fillna('')
business_info['description'] = business_info['description'].fillna('')
business_info['MISC'] = business_info['MISC'].fillna('')
business_info['price'] = business_info['price'].fillna('')
```

```python
# Creating a categories set that contains all unique categories present in the category column of the business_info
categories = set()
def unique_cat(val):
    for i in val:
        categories.add(i)
business_info['category'].apply(unique_cat)
```

```python
# Creating a misc_keys set that contains all unique miscellaneous information present in the misc column of the business_info
misc_keys = set()
def get_keys(val):
    for i in val:
        misc_keys.add(i)
business_info['MISC'].apply(get_keys)
```

```python
misc_dict = {}
for key in misc_keys:
    misc_dict[key] = set()

# Function to extract miscellaneous information and populate misc_dict
def find_misc(val):
    if isinstance(val, str):
        return
    for key, value in val.items():
        if isinstance(value, str):
            misc_dict[key].add(value)
        else:
            for item in value:
                misc_dict[key].add(item)

# Calling the function
business_info['MISC'].apply(find_misc)
```

```python
# Extracts unique values of price for further processing 
prices = business_info['price'].unique()
prices
```

```python
# Assigning tags to price variable for easier accesibility and data processing
def return_price(val):
    if val == '':
        return ''
    if len(val) == 1:
        return 'Cheap'
    if len(val) == 2:
        return 'Budget Friendly'
    if len(val) == 3:
        return 'Expensive'
    return 'Very Expensive'
business_info['price'] = business_info['price'].apply(return_price)
```

```python
# Selecting the miscellaneous information of interest
accepted_misc_keys = ['Health & safety', 'Planning', 'Health and safety', 'Service options', 'Highlights']
```

```python
# Creates a 'content' column with None values and add it in the dataset
business_info['content'] = None

# Function to merge information of description, category, price and misc columns
def merge_cols(row):

    # Adds description
    res = row['description']

    # Adds price if it's not empty
    if row['price'] != '':
        res += ', ' + row['price']

    # Adds each category
    for category in row['category']:
        res += ', ' + category

    # Removes leading comma and space if present
    if res[:2] == ', ':
        res = res[2:]

    # Adds MISC information
    if isinstance(row['MISC'], str):
        return res
    for key in row['MISC']:
        if key not in accepted_misc_keys:
            continue
        if isinstance(row['MISC'][key], str):
            res += ', ' + row['MISC'][key]
        for value in row['MISC'][key]:
            res += ', ' + value
    return res    

# Calling the merge_cols function to update the 'content' column
business_info['content'] = business_info.apply(lambda row: merge_cols(row), axis=1)
```

```python
# Now dropping the 'description', 'category', 'price', and 'MISC' as this information is already stored in content column
business_info.drop(axis=0, inplace=True, columns=['description', 'category', 'price', 'MISC'])
```
